```python
#Pop
pop_beta = {}
pop_error = {}
#LS
ls_beta = {}
ls_error = {}
for n in name_options:
    try:
        with open('results/results_'+n+'.pickle', 'rb') as handle: 
            results = pickle.load(handle)
    except:
        break
    (beta, error) =results['Pop']['Beta +- error']
    pop_beta[n] = beta
    pop_error[n] = error
    k =results['LS']
    ls_beta[n] = k[0]
    ls_error[n] = k[1]
```
