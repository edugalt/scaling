# Creating fitting file

## NSW

```python
NSW = pd.read_pickle(saving_path+"/NSW_SUA_SA2_LGA24Oct.pkl")
NSW
```

```python
df = NSW[['SUA_SA2_LGA_NAME', 'population', 'cases']]
df.head()
```

```python
df['Total population'] = df.groupby(['SUA_SA2_LGA_NAME'])['population',].transform('sum')
df['Total cases'] = df.groupby(['SUA_SA2_LGA_NAME'])['cases',].transform('sum')
```

```python
new_df = df.drop_duplicates(subset=['SUA_SA2_LGA_NAME'])
new_df = new_df.drop(['population', 'cases'], axis=1)
new_df = new_df.rename({'SUA_SA2_LGA_NAME':'SUA_SA2_LGA_NAME', 
                        'Total population': 'population', 'Total cases': 'cases'},
                       axis='columns')
new_df
```

```python
new_df['population'] = new_df['population'].astype(np.int)
frame=pd.DataFrame(new_df.values, columns = ["#SUA_SA2_LGA_NAME", "population", "cases"])
frame.to_csv(txt_path+'covid19_NSW.txt', index=None, sep=',', mode='a')
```

```python
new_df.to_csv(saving_path+'/covid19_NSW.csv' , header=['cities', 'population', 'cases'], 
                 index = False)
```

# USA

```python
USA = pd.read_csv(saving_path+"/USA14Oct.csv")
USA
```

```python
USA['county-state'] = np.nan
USA['county-state'] = USA['county'] + '-' + USA['state']
```

```python
USA = USA.drop(columns=['county','state'])
USA = USA[['county-state','population','cases']]
USA
```

```python
USA['county-state'] = USA['county-state'].astype(np.str)
USA['population'] = USA['population'].astype(np.int)
USA['cases'] = USA['cases'].astype(np.int)
```

```python
USA.reset_index(drop=True)
```

```python
new_USA = USA.copy()
frame=pd.DataFrame(new_USA.values, columns = ["#county", "population", "cases"])
frame.to_csv(txt_path+'covid19_USA.txt', index=None, sep=',', mode='a')

USA.to_csv(saving_path+'covid19_USA.csv' , header=['cities', 'population', 'cases'], 
                 index = False)
```
